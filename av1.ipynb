## Definições: 
### HDI Rank (Classificação do Índice de Desenvolvimento Humano):
O HDI Rank fornece uma classificação numérica para cada país com base em sua pontuação no Índice de Desenvolvimento Humano (HDI). Essa classificação oferece uma visão comparativa do nível de desenvolvimento de um país em relação a outros. As classificações mais baixas do IDH estão associadas a níveis mais altos de desenvolvimento humano, indicando que os países com classificações mais baixas geralmente têm melhores padrões de vida, educação e expectativa de vida em comparação com aqueles com classificações mais altas. 
(Referência: https://hdr.undp.org/data-center/human-development-index#/indicies/HDI)
### Classificação GII (Classificação do Índice de Desigualdade de Gênero):
O GII Rank atribui uma posição numérica a cada país com base em sua pontuação no Índice de Desigualdade de Gênero (GII). Essa classificação oferece insights sobre os esforços de um país para lidar com as disparidades baseadas em gênero. Classificações mais baixas do GII significam menos desigualdade de gênero, indicando que os países com classificações mais baixas tendem a ter oportunidades e direitos mais equitativos para ambos os sexos.
(Referência: https://www.wipo.int/global_innovation_index/en/)
### GII (Índice de Desigualdade de Gênero):
O GII é um índice composto que avalia as desigualdades baseadas em gênero em três dimensões críticas: saúde, empoderamento e participação na força de trabalho. Ele avalia fatores como taxas de mortalidade materna, taxas de natalidade de adolescentes, representação política das mulheres e níveis de educação entre homens e mulheres. Os valores do GII variam de 0 a 1, sendo que valores mais baixos representam menor desigualdade de gênero. Um GII mais baixo reflete um acesso mais equitativo a recursos e oportunidades para todos os gêneros em um país.
(Referência:https://hdr.undp.org/data-center/thematic-composite-indices/gender-inequality-index#/indicies/GII)

```python
import pandas as pd
 
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# Importar os dados de meu repositório no Github
raw = pd.read_csv("https://raw.githubusercontent.com/elinarindle/Python/main/Gender%20Inequality%20Index.csv", header=0)
data = pd.DataFrame(raw)
data.head()
```

## Parte 1: Índice de Desigualdade de Gênero por Continente
Nesta seção, nosso objetivo é criar uma tabela agrupada e, em seguida, gerar um gráfico que ilustra a evolução do Índice de Desigualdade de Gênero médio em cada país ao longo dos anos deste conjunto de dados, que abrange o período de 1990 a 2021.

```python
# Lista de colunas a serem excluídas do DataFrame
columns_to_exclude = ['ISO3', 'Country', 'Hemisphere', 'Human Development Groups', 'UNDP Developing Regions', 'HDI Rank (2021)', 'GII Rank (2021)']

# Remover as colunas especificadas do DataFrame original e criar um novo DataFrame
df = data.drop(columns=columns_to_exclude)

# Exibir as primeiras linhas do novo DataFrame
df.head()
```

```python
# Derreter o DataFrame para converter colunas em linhas
melted_df = pd.melt(df, id_vars=['Continent'], var_name='Year', value_name='Gender Inequality Index')

# Extrair o ano da coluna 'Ano' removendo 'Índice de Desigualdade de Gênero ('
melted_df['Year'] = melted_df['Year'].str.extract(r'(\d{4})').astype(int)

# Remover duplicatas calculando a média
melted_df = melted_df.groupby(['Year', 'Continent'])['Gender Inequality Index'].mean().reset_index()

# Girar o DataFrame para tornar os continentes como colunas
# Agora, pivoted_df possui anos como linhas e continentes como colunas com colunas renomeadas
continents_df = melted_df.pivot(index='Year', columns='Continent', values='Gender Inequality Index')

continents_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar o estilo do Seaborn 
sns.set(style="whitegrid")

# Criar uma figura para o gráfico com tamanho personalizado
plt.figure(figsize=(10, 6))

# Criar um gráfico de linha usando Seaborn com linhas sólidas
sns.lineplot(data=continents_df, dashes=False)

# Definir o rótulo do eixo X
plt.xlabel('Ano')

# Definir o rótulo do eixo Y
plt.ylabel('Índice de Desigualdade de Gênero')

# Definir o título do gráfico
plt.title('Índice de Desigualdade de Gênero ao Longo dos Anos por Continente')

# Adicionar uma legenda ao gráfico com o título 'Continente'
plt.legend(title='Continente')

# Habilitar a grade no gráfico
plt.grid(True)

# Exibir o gráfico
plt.show()
```

### Análise
A partir do gráfico de linhas acima, é evidente que, ao longo do tempo, a desigualdade de gênero diminuiu em todos os continentes. No entanto, os pontos de partida e o ritmo dessa queda variam significativamente. Pegue a Europa, por exemplo, que começou com uma média excepcionalmente baixa no GII de 0,3 em 1990, um nível que nenhum dos outros continentes sequer se aproximou até 2021. Nos últimos anos, a desigualdade de gênero na Europa caiu para quase 0,1, indicando um progresso substancial.

Agora, voltando nossa atenção para continentes que enfrentam desafios mais pronunciados, observamos uma queda consistente na África também. Essa queda é quase paralela à trajetória da Europa, mas começou em um patamar muito mais alto, ultrapassando 0,65 em 1990 e atingindo apenas 0,52 em 2021. Esses números ilustram claramente a jornada significativa que a África ainda tem pela frente para alcançar um nível comparável ao da Europa, apesar de estar progredindo a uma taxa semelhante.

Tanto a América quanto a Ásia também experimentaram uma queda constante na desigualdade de gênero, e suas pontuações médias de GII são bastante próximas. Vale ressaltar que, em 2005, a Ásia ultrapassou a América, apesar de começar com um GII ligeiramente mais alto. Isso indica que a desigualdade de gênero na Ásia diminuiu a uma taxa ligeiramente mais rápida do que a da América.

A história da Oceania é mais complexa, com seu gráfico de linha exibindo flutuações mais visíveis ao longo de períodos mais longos e um aumento recente na desigualdade a partir de 2020. Essas flutuações sugerem que a Oceania enfrentou desafios únicos e pode exigir esforços direcionados para estabilizar e reduzir consistentemente a desigualdade de gênero.

Em resumo, o gráfico de linhas revela que, embora a desigualdade de gênero tenha diminuído globalmente, os pontos de partida, taxas de mudança e trajetórias únicas de cada continente contam uma história complexa de progresso e desafios na busca pela igualdade de gênero. Isso destaca a necessidade de estratégias específicas para cada região para abordar de forma eficaz as disparidades de gênero.

## Parte 2: Classificação dos índices com base na presença de grupos de desenvolvimento humano

Nesta seção, exploramos a distribição como os países são classificados, com base no número de grupos de desenvolvimento humano que possuem. Os índices em análise são o Índice de Desenvolvimento Humano (IDH) e o Índice de Desigualdade de Gênero (GII) do ano de 2021, pois estão disponíveis nesse conjunto de dados. A ideia central é entender como as classificações desses índices variam se a presença de grupos de desenvolvimento humano for classificada como "Baixa", "Média", "Alta" e "Muito alta".

```python
# Lista de colunas a serem excluídas do DataFrame
columns_to_delete = ['ISO3', 'Country', 'Continent', 'Hemisphere', 'UNDP Developing Regions']

# Remover as colunas especificadas do DataFrame original e criar um novo DataFrame
df2 = data.drop(columns=columns_to_delete)

# Exibir as primeiras linhas do novo DataFrame
df2.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crie um único gráfico com dois subgráficos na mesma linha
fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # 1 row, 2 columns

# Lista de colunas para plotagem
columns_to_plot = ['GII Rank (2021)', 'HDI Rank (2021)']

# Loop pelas colunas e crie gráficos de violino
for i, column in enumerate(columns_to_plot):
    sns.violinplot(x='Human Development Groups', y=column, data=df2, ax=axes[i])
    axes[i].set_xlabel(' ')
    axes[i].set_ylabel(column)
    axes[i].set_title(f'{column} por Grupos de Desenvolvimento Humano')

# Ajuste o layout para evitar sobreposição
plt.tight_layout()

# Mostre os gráficos
plt.show()
```

### Análise
Sem dúvida, ambos os gráficos deixam claro que quanto maior a presença dos Grupos de Desenvolvimento Humano, menor é a classificação, o que significa que países com uma presença mais forte desses grupos tendem a ter menos desigualdade de gênero e níveis mais altos de desenvolvimento humano. Essa relação destaca o impacto positivo de promover iniciativas de desenvolvimento humano na redução das disparidades de gênero.

No entanto, ao examinar de perto os gráficos, algumas nuances interessantes surgem. A Classificação do Índice de Desigualdade de Gênero (GII) apresenta uma distribuição muito mais ampla, indicando uma maior variação na forma como os países são classificados em termos de desigualdade de gênero. Isso sugere que existe uma diversidade mais significativa de desafios relacionados ao gênero em diferentes nações.

Por outro lado, a Classificação do Índice de Desenvolvimento Humano (HDI) parece seguir um padrão mais distinto, com as classificações movendo-se em etapas discerníveis de "Baixo" para "Médio" e assim por diante. Isso sugere que os países tendem a se agrupar mais distintamente dentro de certas categorias de desenvolvimento, refletindo a presença de limiares bem definidos nas conquistas de desenvolvimento humano. Além disso, as distribuições do HDI Rank parecem mais simétricas, indicando uma distribuição mais equilibrada de países em diferentes níveis de desenvolvimento.

Em contrapartida, as distribuições do GII Rank mostram mais variabilidade, com valores atípicos estendendo o comprimento dos "violinos" e fazendo com que eles se sobreponham mais do que no caso do HDI Rank. Isso indica que existem países com desempenho excepcional em desigualdade de gênero, tanto melhores quanto piores do que a mediana, contribuindo para uma distribuição menos uniforme das classificações.

Resumindo, enquanto ambos os índices destacam a importância dos Grupos de Desenvolvimento Humano na redução da desigualdade de gênero, o GII Rank exibe uma maior variabilidade e valores atípicos, enquanto o HDI Rank segue um padrão de distribuição mais estruturado e simétrico, revelando diferentes facetas da relação entre desenvolvimento e igualdade de gênero. Essas informações podem orientar intervenções e políticas direcionadas para abordar as disparidades de gênero em escala global.
